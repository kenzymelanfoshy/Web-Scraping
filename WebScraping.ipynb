# Web scraping trail 2

```python
import requests
from bs4 import BeautifulSoup

# Target URL
url = "https://www.egypttoday.com/"

# Fake a browser with headers
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 \
                   (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}

# Request the page
response = requests.get(url, headers=headers)

# Parse HTML
soup = BeautifulSoup(response.content, "lxml")

# Find all Sitecontainer divs
site_containers = soup.find_all("div", class_="Sitecontainer")
print(site_containers)
```

```python
# Print each container's content
for i, container in enumerate(site_containers, start=1):
    print(f"\n📦 Sitecontainer #{i}")
    print("=" * 50)
    print(container.get_text(strip=True, separator='\n'))
```

```python
# Step 5: Extract text from each container
data = []
for i, container in enumerate(site_containers, start=1):
    text = container.get_text(separator='\n', strip=True)
    data.append({"Container_Number": i, "Content": text})

# Step 6: Create DataFrame
df = pd.DataFrame(data)

# Step 7: Save to CSV
df.to_csv("egypttoday_sitecontainers.csv", index=False, encoding='utf-8-sig')

print("✅ Data saved to 'egypttoday_sitecontainers.csv'")
```

# Web Scraping on wikipedia


```python
from bs4 import BeautifulSoup
import requests
```

```python
url ="https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find_all('table')[0]
```

```python
print(table)
```

Separate of th tags

```python
word_titles = table.find_all('th')
```

```python
word_titles
```

```python
word_table_titles = [title.text.strip() for title in word_titles]
print(word_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=word_table_titles)
df
```

"tr" >>> cloumns -----
"td" >>>> rows (data)

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    if len(individual_row_data) == len(df.columns):
        df.loc[len(df)] = individual_row_data

# Optional: display first few rows
print(df.head())
```
